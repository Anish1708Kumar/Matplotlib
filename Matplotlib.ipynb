Answer to Question number 1.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 4 * np.pi, 400)

y1 = np.sin(x)

y2 = np.cos(x / 2)

plt.plot(x, y1, color='blue', linestyle='solid', label='sin(x)')
plt.plot(x, y2, color='red', linestyle='dashed', label='cos(x/2)')

plt.xlabel('x')
plt.ylabel('y')

plt.title('Sin vs. Cos/2')

plt.legend()

plt.savefig('sin_vs_cos2.png')

plt.show()
```

Answer to Question number 2.

```python
import pandas as pd
import matplotlib.pyplot as plt

url = 'https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv'
data = pd.read_csv(url)

grouped_data = data.groupby('species')
mean_sepal_lengths = grouped_data['sepal_length'].mean()

species = list(mean_sepal_lengths.index)
means = list(mean_sepal_lengths.values)

plt.bar(species, means, color='skyblue')

for i in range(len(species)):
    value = round(means[i], 2)
    plt.text(i, means[i] + 0.05, str(value), ha='center')

plt.xlabel('Species')
plt.ylabel('Mean Sepal Length')
plt.title('Mean Sepal Length of Each Iris Species')

plt.show()
```

Answer to Question number 3.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

spot_url = "https://raw.githubusercontent.com/CarlHatoum/Spotify-TOP-50-songs/master/top50.csv"
spotify_data = pd.read_csv(spot_url, encoding='latin1')

pair_data = spotify_data[['Danceability', 'Energy', 'Valence.', 'Speechiness.']]
pair_data['Genre'] = spotify_data['Genre']

sns.set_style("darkgrid")

sns.pairplot(pair_data, hue='Genre')

plt.suptitle("Spotify Pairplot by Genre", y=1.02)
plt.tight_layout()
plt.savefig("q3_spotify_pairplot.png")
plt.show()
```

Answer to Question number 4.

```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(0)
A = np.random.normal(0, 1, 1000)
B = np.random.normal(2, 1.5, 1000)

fig, axs = plt.subplots(2, 1, figsize=(8, 8))
fig.suptitle("Comparing Two Normal Distributions")

axs[0].hist(A, bins=30, alpha=0.6, label='A ~ N(0,1)')
axs[0].hist(B, bins=30, alpha=0.6, label='B ~ N(2,1.5)')
axs[0].legend()
axs[0].set_title("Overlapping Histograms")

axs[1].boxplot([A, B], labels=['A', 'B'])
axs[1].set_title("Box and Whisker Plots")

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```
